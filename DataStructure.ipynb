# Data Structure ->
### They are used for storing the multiple values together.

## Beyond single values ->
### So, far we have stored only one value per variable. But what if you need to store multiple values?
# That's where data structure comes in

# Think of data structure as containers ->
## 1. List => Like a shopping list (ordered items)
## 2. Dictonaries => Lika a phone book (name > number) such as key value pairs
## 3. Tuples => Loke coordinates (fixed value)
## 4. Sets => Like a bog of unique items.

# Q. Which one to use when ?
1. Lists => When order of the matters and you need to change the items.
2. Dictonaires => When you need to look up values by name.
3. Tuples => When data shouldn't change(like coordinates)
4. Sets => When you only care about unique values

```python
# Lists -> Works with the ordered collections
# Q.What are lists?
# Ans. Lists are python's most versatile data structure.They are like containers that can hold multiple items in a specific orders.
# Think of a list like =>
# 1 A shopping list(milk,eggs,bread)
# 2 A to-do (tasks in order)
# 3 A playlist

# Creating a list ->
# my_list = []

# :ist with items =>
fruits = ["apple","mango","banana"]
numbers = [1,2,3,4,5,6,7]
mixed = ["hello",42,True,3.14]
print(mixed)
mixed[2]

# Accessing items =>
needs = ["roti","kapda","makan"]
# Get itmes =>
print(needs[0])
print(needs[1])
print(needs[-1]) # this will give the last item.
print(needs[-2]) # this will give the second last item.

first = needs[0] # this will print the first element of the list with the new variable which is storing the value.
print(first)

# We can also change the value of the list elements ->

protein = ["MB","Avatar","Atom"]
# Change the item ->
protein[0] = "MuscleTech"
print(protein) # now the value of the 0th index has been changed.

# Add items ->
protein.append("FuelOne") # this will end the element at the last position
protein.insert(1,"GNC") # this will print the element at the 1st index
print(protein)

# To remove the items ->
protein.remove("MuscleTech") # this will remove the element by the value.
print(protein)
last = protein.pop() # remove and return the last.
print(protein)
del protein[0] # this will remove the index
print(protein)
```

```python
# Methods of List ->
numbers = [1,2,3,4,5,5,6,1,2,3]
# Information :-
print(len(numbers)) # this will print the length of the list.
print(numbers.count(1)) # the count is used for counting the occurence of the element.
print(numbers.index(4)) # this is to find the value of the index finding the position of the element.

# Sorting ->
numbers.sort()
print(numbers) # this will sort the numbers in the ascending to descending.

numbers.reverse() # Reverse the order of the list.
print(numbers) # this will print in the form of descending to ascending.

# Copy ->
new_numbers = numbers.copy()
print(new_numbers) # this will copy the list in the new list.
```

# Dictionaries 🥰
### Store the data with the key-value pairs.



```python
# Q. What are dictionaries?
# Ans. Dictionaries store data in key-value pairs. Think of them like a real dictionary where you look up a word(key) to find its definition(value)
# Some real world example ->
# Phone book : name > phone number
# Menu : dish > price
# User profile : username > user info

# To create a dictionary ->
my_dict = {} # this is the empty dictionary.
# Dictionary with data ->
person = {
    "name":"Aditya",
    "age":23,
    "city": "Gwalior"
}
print(person)

# Different ways to create ->
scores = dict(math=95, english=87,science=92)
print(scores)

# Accessing the value ->
person1 = {"name":"Alice","age":23,"city":"gwalior"}
# Get the value by key ->
print(person1["name"]) # This is the key for the value.
print(person1["age"])
# can also add values ->
person["licence"] = True
print(person)
# can also delete the elements from the dictionary.
del person["age"]
print(person)
# Safer with the get() :-
print(person1.get("job")) #No errors will print None
print(person1.get("job","Unknown"))

# Changing dictionaries ->
person2 = {"name":"Aditya","age":25}
# Add or update ->
person2["email"] = "soni@gmail.com" # this will add the new item in the dict.
person2["age"] = 23 # this will update the given value of the age
print(person2)

# Remove the items ->
del person2["age"] # this will delete the whole item with its value.
print(person2)
name = person2.pop("name") # remove or return the value
person2.clear() # this will remove the whole items in the dict.
print(person2)

# Dictionary methods =>

info = {"name":"aditya","age":23,"city":"gwl"}
# To get all keys, values or items ->
print(info.keys()) # this will give the list of the keys in the dict.
print(info.values()) # this will print the list of the value of the keys
print(info.items()) # this will give everything

# Check if key exists :-
if"name"in info:
  print("Name found")
else:
  print("Not found")

# Update multiple values :-
info.update({"age":22,"job":"Berozgaar"})
print(info)

# Nested Dictionaries -> It is called as dictionary of dictionaries
students={
    "Alice":{"age":23,"grade":"A"},
    "bob":{"age":21,"grade":"B"},
    "charlie":{"age":19,"grade":"C"}
}

# Accessing the nested dict ->
print(students["Alice"]["grade"])
```

# Tuples ->
### Tuples are just like lists, but they can't be changed once created.They are immutable(unchangeable) sequences.

```python
# Use tuples for data that shouldn't change:
# Coordinates(x,y)
# RGB colors(255,0,0)
# Database records
# Function return values

# Creating tuples ->
# Empty tuple->
empty = ()

# Tuple with items ->
point = (3,5)
colors = ("red","green","blue")
print(colors)

# Single item tuple needs comma!
single = (42,) # Note the comma
not_tuple = (42) # This is just the 42 in the parantheses

# Without parantheses (implicit)
coordinates = 10,20
print(coordinates)

# Accessing items => Just like lists,tuples use indexing:
point = (2,5)
colors1 = ("violet","purple","pink")

# Get items =>
print(point[0]) # 2
print(colors1[-1]) # pink

# Slicing wokrs too ->
print(colors1[0:2])
print(colors1)
```
