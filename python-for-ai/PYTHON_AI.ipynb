# Operators ->
### Math,comparison and logic in Python

```python
# Q. What are opartors ?
# Ans. Operators are symbols that perform operations on values. Think of them as the "verbs of the programming"- they make things happen!
# We already knows most of them ->
# 1. Calculate: - +,-,*,/
# 2. Compare: ==, >, <
# 3. Combie: and, or , not

# Arthemtic operations -> These work just like calculator.
# Basic math :-

print(10.4+8)
print(12-11)
print(10*3)
print(10/3)

# Special operations ->
print(10 //3)
print(10%3)
print(10 ** 3)

# Order of operations - Python follows main rules (PEMDAS) => Parentheses, Exponents, Multiplication and Division (left to right), and Addition and Subtraction (left to right)
print(2+3*4) # the answer will be 14 not 20 cause in python we do not focus on the bodmas rule like first we will solve bracket not that bs.
print((2+3)*4) # now this will be 20 cause it follows BODMAS.

# Comparison operations ->
age = 24
print(age == 24)
print(age!=24)
print(age > 18)
print(age < 25)
print(age >=24)
print(age <= 24)

# Logical operators => These combine boolean values and conditions.
age = 25
has_license=True

# AND - both must be true ->
can_drive = age >=18 and has_license
print(can_drive)

# OR - atleast one must be true ->
day = "Friday"
is_week = day == "Saturday" or "Friday"
print(is_week)

# NOT - reverse the values ->
is_adult = age >=18
is_child = not is_adult
print(is_child)

# Truth tables => Understanding how and,or , and not work :-
# AND -> Both must be True
print(True and True)
print(True and False)
print(False and False)

# OR -> At least one must be True
print(True or False)
print(False or False)

# Not => flips the values -
print(not True) # this will become false
print(not False) # this will become true

# Assignment Shortcuts :-
score = score +10
# we can write ->
score += 10 # both are same operations only the syntax has changed.

# Works with all the opeators
x = 10
x += 5
print(x) # 15
x -= 4
print(x) # 11

# Common mistakes :-
# Wrong ->
average = 10+20+30 / 3 #40.0
# Right ->
average = (10+20+30)/3 #20.0
```

# String manipulations ->
### Advanced text operations


```python
# Working with text :-
# Now that we know how to create strings and what strings are let's learn how to manipulate them.Python gives you powerful tools to work with text.

# String operations :-

# 1. String concatination (joining) ->
# Using + old fashioned
first_name = "Aditya"
last_name = "Soni"
full_name = first_name + " " + last_name
print(full_name)

# Using f-strings modern python way
greet = f"Hello, {first_name}"
print(greet)

# Multiple variables :-
age = 23
intro = f"I am {first_name} and I am {age} years old"
print(intro)

# Repitions -> this will just multiply the value which you will pass.
star = "*"
stars = star * 10
print(stars)

# String methods -> Python strings come with many built-in methods \n
# functions you can call directly and clean up any string . These methods
# let you transform text, search for patterns and clean up data. The
# best part is we can often guess what they do from their names
# upper() makes text uppercase, replace() replace text and so on.

# changing Case =>
text= "Aditya Soni"
print(text.upper())
print(text.lower())
print(text.title())

# Cleaning Strings :-
messy = "   hello world       "
print(messy.strip()) # this will remove the spaces which are white spaces but not the tabs one

price = "$12.22"
print(price.strip("2")) # this will remove the after decimal value.

tags = "@aditya"
print(tags.strip("@"))

# Finding and replacing ->
message = "I love python language"

# Check if something exists ->
print("love" in message) # cause it is not in my real life.
print(message.startswith("I"))
print(message.endswith("I"))

# Find position ->
print(message.find("love")) # this will tell the first occurence like the indices of the value.
print(message.count("language"))

# Replace ->
new_message = message.replace("love","hate") # this will change the value of love to the hate after finding it in the string message/
print(new_message)
```

# If Statements ->
### Make decisions in your code

```python
# Q. What are if statements?
# Ans. If statements let your program make decisions.They check is something is true or false, then act accordingly.
# Real life logic :-
#  IF it is raining THEN take umbrella
#  IF battery < 20% THEN charge phone
#  IF password correct THEN allow access.

temp = 18
if temp<18:
  print(f"The weather is hot {temp}")
else:
  print(f"The weather is cold")
print(temp)
```

```python
# We can use multiple elif ->
score = 70
if score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
else:
  print("D")
```

```python
# Now multiple conditions ->
age = 23
has_license = True
weekend = True
holiday=False
raining = False

# Both must be true ->
if age >= 18 and has_license:
  print("You are good")
# Atleast one must be true ->
if weekend or holiday:
  print("No worl today")
# Reverse the if condition ->
if not raining:
  print("Let's go outside")
```

```python
# Nested if statement ->
has_ticket = True
age = 50

if has_ticket:
  if age >= 18:
    print("Enjoy the show")
  else:
    print("Need supervision")
else:
  print("Buy a ticket")
```
